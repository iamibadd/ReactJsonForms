```python
def format_hierarchy(hierarchy):
    # Transform flat hierarchy list into a nested structure or a readable format
    hierarchy_dict = {}
    for child_id, parent_id in hierarchy:
        hierarchy_dict[child_id] = parent_id

    def construct_path(child_id):
        path = [child_id]
        while child_id in hierarchy_dict and hierarchy_dict[child_id]:
            child_id = hierarchy_dict[child_id]
            path.append(child_id)
        return " > ".join(reversed(path))

    return [construct_path(child_id) for child_id, _ in hierarchy]

hierarchy = functions.fetch_object_hierarchy(conn)
formatted_hierarchy = format_hierarchy(hierarchy)
print("Object Hierarchy:")
for path in formatted_hierarchy:
    print(path)
```
